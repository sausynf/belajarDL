## B. Pengenalan ANN

<p>Jaringan saraf tiruan, juga dikenal sebagai artificial neural networks, adalah model komputasi yang didasarkan pada struktur dan fungsi otak manusia. Ini terdiri dari neuron, unit kecil yang saling terhubung dengan bobot yang dapat diubah.</p>

<p>Neural network digunakan untuk memproses data, melakukan tugas seperti pengenalan pola, klasifikasi, dan regresi, serta mempelajari pola-pola ini dari data pelatihan.</p>

<p>Lapisan-lapisan neuron memungkinkan neural network untuk mempelajari representasi fitur data yang kompleks. Ini adalah salah satu teknik penting dalam deep learning, yang menggunakan artificial neural networks yang dalam dan kompleks untuk menyelesaikan masalah yang kompleks.</p>

<h3>Komponen Utama Artificial Neural Networks</h3>
<ol>
    <li><b>Neuron</b> <br>
        - Unit dasar dalam ANN yang menerima input, melakukan perhitungan, dan menghasilkan output.<br>
        - Input dikalikan dengan bobot yang sesuai dan dijumlahkan.<br>
        - Hasilnya kemudian diteruskan melalui fungsi aktivasi.</li>
    <br>
    <li><b>Layer (Lapisan) </b><br>
        - Neuron dikelompokkan dalam 3 lapisan yang berbeda: lapisan input, lapisan tersembunyi (hidden), dan lapisan output.<br>
        - ANN terdiri dari satu atau lebih lapisan neuron.<br>
        - Lapisan input : menerima data masukan<br>
        - Lapisan tersembunyi : Membuat representasi fitur internal.<br>
        - Lapisan output : Menghasilkan output terakhir</li>
    <br>
    <li><b>Weights (Bobot)</b> <br>
    - Bobot menghubungkan neuron dalam lapisan yang berbeda.<br>
    - Mereka mengontrol seberapa kuat pengaruh suatu neuron terhadap neuron lain.<br>
    - Bobot diubah selama proses pelatihan jaringan.</li>
    <br>
    <li><b>Activation Function (Fungsi Aktivasi)</b> <br>
    - Fungsi yang diterapkan pada output neuron untuk menentukan apakah mereka harus aktif atau tidak.<br>
    - Contoh fungsi aktivasi adalah ReLU, sigmoid, dan tanh.<br>
    - Mereka memperkenalkan non-linearitas ke dalam model.</li>
<ol>


<h3>Cara Artificial Neural Networks Bekerja</h3>

Untitled Diagram-Page-4.drawio.png

<p><b>Penjelasan :</b></p>

<ol>
    <li><b>Feedforward</b><br>
    Ini merupakan hal pertama yang dilakukan dalam training ANN, Pada proses ini terdiri dari 3 tahap:<br>
        <b>- Input layer :</b><br>  Layer ini menerima input data. Setiap neuron dalam layer ini mewakili satu fitur dari data.<br>
        <b>- Hidden layer :</b><br>  Layer ini berada di antara input dan output layer. Setiap neuron dalam layer ini menerima input, mengalikan dengan bobot, menambahkan bias, dan meneruskan hasil melalui fungsi aktivasi.<br>
    <b>- Output layer :</b><br>   Layer ini memberikan hasil akhir dari neural network. Setiap neuron dalam layer ini juga menerima input, mengalikan dengan bobot, menambahkan bias, dan meneruskan hasil melalui fungsi aktivasi.</li><br>
    <li><b>Hitung Error</b><br>
     Setelah feedforward, kita menghitung error dari hasil output layer dengan menghitung selisihnya dengan output yang sebenarnya.</li><br>
    <li><b>Backpropagation</b><br>
        <b>- Apa itu Backpropagation?</b><br>
        Backpropagation adalah metode yang digunakan dalam neural network untuk menyesuaikan bobot dan bias berdasarkan error atau cost yang dihasilkan oleh output layer. Backpropagation adalah tulang punggung dari pelatihan neural network.<br>
        <b>- Cara kerja</b> <br>
        Setelah kita mendapatkan error, kita mulai backward pass. Di sini, kita menghitung gradien dari error terhadap bobot dan bias di setiap layer, mulai dari output layer dan bergerak mundur ke input layer</li><br>
    <li><b>Algoritma Optimasi</b><br>
    Setelah kita memiliki gradien, kita dapat menggunakan mereka untuk memperbarui bobot dan bias menggunakan algoritma optimasi. Ini biasanya dilakukan dengan mengurangi gradien yang dihitung dari bobot dan bias saat ini, dikalikan dengan learning rate.<br>
        <b>Untuk apa algoritma optimasi?</b><br>
    Algoritma optimasi digunakan dalam training ANN untuk memperbarui bobot dan bias dengan tujuan meminimalkan fungsi cost.</li>
</ol>

<p><b>Macam-macam Activation Functions</b><p>
<p>Fungsi aktivasi memainkan peran penting dalam arsitektur Neural Network. Fungsi ini bertugas untuk menentukan output neuron berdasarkan set input tertentu. </p>
<p> Fungsi ini memperkenalkan non-linearitas ke dalam model, yang memungkinkan jaringan untuk memodelkan hubungan yang lebih kompleks dalam data.</p>

<p><b>1. Sigmoid</b></p>
<p>Fungsi aktivasi sigmoid mengubah input menjadi nilai antara 0 dan 1. Ini berguna dalam kasus klasifikasi biner.</p>
<br>
<p><b>Keuntungan :</b></p>
<p>Output berkisar antara 0 dan 1, berguna untuk mengubah nilai menjadi probabilitas.</p>
<p><b>Kekurangan :</b></p>
<p>Sebesar dan sekecil apapun nilai input, Output yang dikeluarkan hanya di sekitar 0 sampai 1.</p>

```python
import numpy as np

# Membuat fungsi Sigmoid
def sigmoid(input):
    return 1 / (1 + np.exp(-input))

# Contoh data input
x = np.array([-2.0, -1.0, 0.0, 1.0, 2.0])

for i in x:
    print("Angka Input : ", i, "| Output Sigmoid : ", sigmoid(i))
```

```python
# Membuat Grafik Output Sigmoid
import matplotlib.pyplot as plt

# Membuat data yang akan dimasukkan ke sigmoid
# Data ini memiliki jumlah sebanyak 100 dengan rentang dari -5 sampai 5
x = np.linspace(-5, 5, 100)

# Menyimpan output sigmoid
output = sigmoid(x)

# Membuat grafik output sigmoid
plt.plot(x, output, label='Output Sigmoid')
plt.title('Sigmoid Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

<p><b>2. ReLU (Rectified Linear Unit)</b></p>
<p>Fungsi aktivasi ReLU mengubah input yang bernilai negatif menjadi 0, sementara input yang bernilai positif tidak diubah.</p>
<br>
<p><b>Keuntungan :</b></p>
<p> ReLU sangat efisien secara komputasi.</p>
<p><b>Kekurangan :</b></p>
<p>ReLU memiliki apa yang dikenal sebagai "dying ReLU problem" - jika neuron yang menggunakan fungsi ReLU mendapatkan input negatif, maka neuron tersebut akan "mati", atau berhenti belajar.</p>

```python
# Membuat fungsi ReLU
def ReLU(input):
    return np.maximum(input, 0)

# Contoh data input
x = np.array([-2.0, -1.0, 0.0, 1.0, 2.0])

for i in x:
    print("Angka Input : ", i, "| Output ReLU : ", ReLU(i))
```

```python
# Membuat Grafik Output ReLU
# Membuat data yang akan dimasukkan ke ReLU
# Data ini memiliki jumlah sebanyak 100 dengan rentang dari -5 sampai 5
x = np.linspace(-5, 5, 100)

# Menyimpan output ReLU
output = ReLU(x)

# Membuat grafik output ReLU
plt.plot(x, output, label='Output ReLU')
plt.title('ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

<p><b>3. Tanh</b></p>
<p>Tanh adalah alternatif lain untuk sigmoid.  Output dari fungsi ini memiliki rentang lebih besar dari sigmoid yaitu berkisar antara -1 hingga 1, yang menjadikannya lebih baik dalam menangani negatif input</p>
<br>
<p><b>Keuntungan :</b></p>
<p>  Sama seperti sigmoid, tetapi lebih baik menangani input negatif.</p>
<p><b>Kekurangan :</b></p>
<p>Sebesar dan sekecil apapun nilai input, Output yang dikeluarkan hanya di sekitar -1 sampai 1.</p>

```python
# Membuat fungsi tanh
def tanh(input):
    return np.tanh(input)

# Contoh data input
x = np.array([-2.0, -1.0, 0.0, 1.0, 2.0])

for i in x:
    print("Angka Input : ", i, "| Output Tanh : ", tanh(i))
```

```python
# Membuat Grafik Output tanh
# Membuat data yang akan dimasukkan ke tanh
# Data ini memiliki jumlah sebanyak 100 dengan rentang dari -5 sampai 5
x = np.linspace(-5, 5, 100)

# Menyimpan output tanh
output = tanh(x)

# Membuat grafik output tanh
plt.plot(x, output, label='Output Tanh')
plt.title('Tanh Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

<p><b>4. Softmax</b></p>
<p>Fungsi aktivasi softmax digunakan dalam lapisan output jaringan yang digunakan untuk klasifikasi multi-kelas. Ini mengubah nilai input menjadi vektor probabilitas.</p>
<br>
<p><b>Keuntungan :</b></p>
<p>   Berguna untuk klasifikasi multi-kelas.</p>
<p><b>Kekurangan :</b></p>
<p> Tidak cocok untuk digunakan di hidden layers.</p>

```python
# Membuat fungsi Softmax
def softmax(input):
    e_x = np.exp(x - np.max(input))
    return e_x / e_x.sum()

# Contoh data input
x = np.array([-2.0, -1.0, 0.0, 1.0, 2.0])

for i in x:
    print("Angka Input : ", i, "| Output Softmax : ", softmax(i))
```

<p><b>5. Leaky ReLU</b></p>
<p>Leaky ReLU adalah varian dari fungsi aktivasi ReLU yang memungkinkan sebagian kecil nilai negatif untuk tetap melewati tanpa menjadi nol mutlak. Ini membantu mengatasi masalah "neuron mati" yang bisa terjadi pada ReLU biasa.</p>
<br>
<p><b>Keuntungan :</b></p>
<p>   Mengatasi masalah dying ReLU.</p>
<p><b>Kekurangan :</b></p>
<p>  Hasilnya mungkin tidak stabil untuk data noise.</p>

```python
# Membuat fungsi Leaky ReLU
def LeakyReLU(input):
    return np.maximum(0.01*input, input)

# Contoh data input
x = np.array([-2.0, -1.0, 0.0, 1.0, 2.0])

for i in x:
    print("Angka Input : ", i, "| Output Leaky ReLU : ", LeakyReLU(i))
```

```python
# Membuat Grafik Output Leaky ReLU
# Membuat data yang akan dimasukkan ke Leaky ReLU
# Data ini memiliki jumlah sebanyak 100 dengan rentang dari -10 sampai 5
x = np.linspace(-10, 5, 100)

# Menyimpan output Leaky ReLU
output = LeakyReLU(x)

# Membuat grafik output Leaky ReLU
plt.plot(x, output, label='Output Leaky ReLU')
plt.title('Leaky ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

<p><b>Macam-macam Cost Functions</b><p>
<p>Dalam konteks Neural Networks, Cost Function, juga dikenal sebagai Loss Function, adalah metrik yang digunakan untuk mengukur sejauh mana prediksi model dari nilai yang sebenarnya. Dengan kata lain, cost function memberikan ukuran "kesalahan" dalam prediksi model kita. </p>

### **1. Mean Squared Error**

Mean Squared Error adalah salah satu fungsi kerugian yang paling umum digunakan dalam masalah regresi. Fungsi ini mengukur seberapa dekat prediksi model dengan nilai sebenarnya dengan cara menghitung rata-rata kuadrat selisih antara prediksi dan nilai sebenarnya.

#### **Rumus:**

$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$

#### **Penjelasan:**

- $MSE$ adalah Mean Squared Error, yang merupakan nilai kerugian yang dihitung.
- $n$ adalah jumlah sampel atau data poin dalam dataset.
- $y_i$ adalah nilai sebenarnya dari data poin ke-$(i)$.
- $\hat{y}_i$ adalah prediksi model untuk data poin ke-$(i)$.
- $\sum$ mengindikasikan penjumlahan dari semua elemen dalam rangkaian.


<p><b>2. Binary Crossentropy</b></p>
<p>Fungsi loss Binary Cross-Entropy, juga dikenal sebagai Log Loss, digunakan dalam masalah klasifikasi biner.</p>
<p>Fungsi ini mengukur seberapa baik model klasifikasi biner memprediksi label kelas yang benar atau salah untuk setiap sampel dalam dataset.</p>
<p>Binary Cross-Entropy biasanya digunakan ketika Anda memiliki dua kelas yang mungkin, seperti "positif" dan "negatif."</p>
<br>
<p><b>Rumus :</b></p>
<p>   $L(y, \hat{y}) = -\frac{1}{n} \sum_{i=1}^{n} [y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i)]$</p>
<p><b>Penjelasan :</b></p>
<ul>
    <li>$L(y, \hat{y})$ adalah nilai kerugian yang dihitung oleh fungsi Binary Cross-Entropy.</li>
    <li>$n$ adalah jumlah sampel dalam dataset.</li>
    <li>$y_i$ adalah label kelas yang benar untuk sampel ke-\(i\).</li>
    <li>\(\hat{y}_i\) adalah prediksi model (probabilitas kelas positif) untuk sampel ke-\(i\).</li>
    <li>\(\log\) adalah logaritma alami (logaritma basis \(e\)).</li>
</ul>
<br>
<p><b>NOTE*:</b></p>
<p>Fungsi loss Binary Crossentropy bisa juga digunakan untuk kasus klasifikasi multi-kelas tetapi digunakan pada data yang memiliki target yang bersifat ambigu atau campuran. Contohnya warna.</p>

```python
# contoh Data Target fungsi loss Binary Crossentropy
# Target Data Hanya mempunyai 2 kelas, 0 dan 1
y = [0, 0, 1, 1, 0, 1, 0]

# contoh Data Target Multi-Kelas yang Menggunakan fungsi loss Binary Crossentropy
y =[
    [0, 0, 0],
    [255, 0, 0],
    [20, 255, 0],
    ....
]

# atau juga bisa bentuknya seperti ini
y = [
  [0, 0.1, 0.2],
  [1, 0, 0.1],
  [1, 0.8, 0],
  ...
]
```

<p><b>3. Sparse Categorical Crossentropy</b></p>
<p>Fungsi loss "sparse categorical crossentropy" adalah jenis fungsi loss yang digunakan dalam masalah klasifikasi multikelas ketika label kelas yang diberikan dalam bentuk bilangan bulat (integer) bukan dalam bentuk one-hot encoding, misalnya label kelas ada 0, 1, 2, 3, dan seterusnya. </p>
<br>
<p><b>Rumus :</b></p>
<p>  $L(y, \hat{y}) = -\frac{1}{n} \sum_{i=1}^{n} \log(\hat{y}_i)$</p>
<p><b>Penjelasan :</b></p>
<ul>
    <li>\(L(y, \hat{y})\) adalah nilai kerugian yang dihitung oleh fungsi <b>sparse categorical crossentropy</b>.</li>
    <li>\(n\) adalah jumlah sampel dalam dataset.</li>
    <li>\(y_i\) adalah label kelas yang benar (dalam bentuk bilangan bulat) untuk sampel ke-\(i\).</li>
    <li>\(\hat{y}_i\) adalah prediksi model (dalam bentuk distribusi probabilitas) untuk sampel ke-\(i\).</li>
    <li>\(\log\) adalah logaritma alami (logaritma basis \(e\)).</li>
</ul>

<p><b>Macam-macam Algoritma Optimasi</b><p>
<p>Algoritma optimasi digunakan dalam pelatihan neural network untuk memperbarui bobot dan bias dengan tujuan meminimalkan fungsi cost. </p><br>
<p><b>1. Stochastic Gradient Descent (SGD)</b></p>
<p>SGD adalah salah satu algoritma optimasi yang paling umum digunakan dalam pelatihan model. Ini adalah varian dari algoritma Gradient Descent yang memperbarui parameter model dengan menghitung gradien dari fungsi kerugian untuk sejumlah kecil sampel (mini-batch) pada setiap iterasi, bukan seluruh dataset pelatihan.</p>
